```python
#for reading csv files, converting data to dataframe for easier manipulaton
import pandas as pd

#to find words freq
from collections import Counter

import warnings
warnings.filterwarnings(action="ignore")
```

```python
review_df=pd.read_csv("processed_data_file.csv")
review_df.head()
```

```python
from textblob import TextBlob

def getPolarity(tweet):
    return TextBlob(str(tweet)).sentiment.polarity

# Define function to obtain Sentiment category
def getSentimentTextBlob(polarity):
    if polarity < -0.3:
        return "Negative"
    elif polarity > 0.3:
        return "Positive"
    else:
        return "Neutral"


textblob_df = review_df.copy()
```

```python
# Apply the functions to respective columns
textblob_df['Polarity']=textblob_df['proceessed_text'].apply(getPolarity)
textblob_df['Sentiment']=textblob_df['Polarity'].apply(getSentimentTextBlob)
textblob_df['Sentiment'].value_counts()
```

```python
df_sorted = textblob_df.sort_values(by='score', ascending=True)
filtered_df = df_sorted[(df_sorted['Sentiment'] =="Neutral") & (textblob_df["Polarity"]>0.0) & (df_sorted["score"]==1)]
filtered_df.head()
```

```python
#rule based sentiment analyzer
from nltk.sentiment.vader import SentimentIntensityAnalyzer
sid = SentimentIntensityAnalyzer()
vader_df = review_df.copy()


def getPolarity_compound_vader(tweet):
    return sid.polarity_scores(str(tweet))["compound"]

def getPolarity_neg_vader(tweet):
    return sid.polarity_scores(str(tweet))["neg"]

def getPolarity_pos_vader(tweet):
    return sid.polarity_scores(str(tweet))["pos"]


# Define function to obtain Sentiment category
def getSentimentTextBlob_vader(polarity):
    if polarity < - 0.05:
        return "Negative"
    elif polarity > 0.05:
        return "Positive"
    else:
        return "Neutral"

```

```python
vader_df['Compound']=vader_df['proceessed_text'].apply(getPolarity_compound_vader)
vader_df["pos"]=vader_df['proceessed_text'].apply(getPolarity_pos_vader)
vader_df["neg"]=vader_df['proceessed_text'].apply(getPolarity_neg_vader)
vader_df['Sentiment']=vader_df['Compound'].apply(getSentimentTextBlob_vader)
vader_df['Sentiment'].value_counts()
```

```python
#custom query builder
custom_query_builder_df = vader_df.copy()
custom_query_builder_df.drop(columns=["text"],inplace=True)

#lets create a month and year data
custom_query_builder_df['at'] = pd.to_datetime(custom_query_builder_df['at'])
custom_query_builder_df.info()
```

```python
star_count=[4,5]
start_date=2022  #default = min
end_date=2024    #default = max
duration_between= 6   #default =6

starwise_sentiment_analysis=""
duration_wise=""

most_important_five_factors=""
```

```python
z=custom_query_builder_df[(custom_query_builder_df["score"].isin(star_count)) & (custom_query_builder_df["at"].dt.year>=start_date) & (custom_query_builder_df["at"].dt.year<=end_date)].assign(year=custom_query_builder_df["at"].dt.year,duration=custom_query_builder_df["at"].apply(lambda date: (date.month - 1) // duration_between + 1)).groupby(["year","duration","score","Sentiment"])[["proceessed_text"]].count()
z=z.rename(columns={"proceessed_text": "count"})
##run this

json_table = z.to_json(orient ='index') 
print(json_table)
```

```python
import json
nested_data = {}
json_table_key=json.loads(json_table)
# Process each key-value pair
for key, value in json_table_key.items():
    # Parse the key into components
    year, month, duration, sentiment = eval(key)
    
    # Ensure the structure exists
    if year not in nested_data:
        nested_data[year] = {}
    if month not in nested_data[year]:
        nested_data[year][month] = {}
    if duration not in nested_data[year][month]:
        nested_data[year][month][duration] = {}
    
    # Add the sentiment data
    nested_data[year][month][duration][sentiment] = value
print(nested_data)
```

```python
# Convert to JSON string for better readability
with open("output_json_format.json", "w+") as outfile:
    json.dump(nested_data, outfile)
    outfile.close()
```

```python
# most important 5 factors 
groups=custom_query_builder_df[(custom_query_builder_df["score"].isin(star_count)) & (custom_query_builder_df["at"].dt.year>=start_date) & (custom_query_builder_df["at"].dt.year<=end_date)].assign(year=custom_query_builder_df["at"].dt.year,duration=custom_query_builder_df["at"].apply(lambda date: (date.month - 1) // duration_between + 1)).groupby(["year","duration","score","Sentiment"])
    
zz = groups[["proceessed_text"]].count().reset_index()
zz = zz.rename(columns={"proceessed_text": "count"})

# Find the most important words for each group
important_words_per_group = {}

for name, group in groups:
    combined_text = "".join(str(group["proceessed_text"]))
    word_freq = Counter(combined_text.split())
    # top_freq_words = [word for word, freq in word_freq.most_common(2)]
    # important_words_per_group[name] = top_freq_words

    top_freq_words = {word.lower(): freq for word, freq in word_freq.most_common(3)}
    important_words_per_group[name] = top_freq_words

# Add the important words to the zz DataFrame
zz["important_words"] = zz.apply(lambda row: important_words_per_group[(row["year"], row["duration"], row["score"], row["Sentiment"])], axis=1)

```

```python
json_table_words_str = zz.to_json(orient ='index') 
print("", json_table_words_str, "\n")
type(json_table_words_str)
```

```python
# Initialize the nested dictionary
nested_data_words = {}
json_table_words=json.loads(json_table_words_str)
# Process each item in the input data
for key, value in json_table_words.items():
    year = value["year"]
    duration = value["duration"]
    score = value["score"]
    sentiment = value["Sentiment"]
    count = value["count"]
    important_words = value["important_words"]

    # Ensure the structure exists
    if year not in nested_data_words:
        nested_data_words[year] = {}
    if duration not in nested_data_words[year]:
        nested_data_words[year][duration] = {}
    if score not in nested_data_words[year][duration]:
        nested_data_words[year][duration][score] = {}
    if sentiment not in nested_data_words[year][duration][score]:
        nested_data_words[year][duration][score][sentiment] = {}
    
    # Add the count and important words
    nested_data_words[year][duration][score][sentiment] = {
        "count": count,
        "important_words": important_words
    }

```

```python
# Convert to JSON string for better readability
json_output_words = json.dumps(nested_data_words)
with open("output_json_format_words.json", "w+") as outfile_words:
    json.dump(nested_data_words, outfile_words)
    outfile_words.close()
```
